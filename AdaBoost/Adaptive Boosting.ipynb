```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
data,y = load_iris(return_X_y=True , as_frame=True)
```

```python
df= pd.concat([data,y], axis=1)
```

```python
df
```

```python
#step 1  - assign sample weights
df['sample_weights'] = 1/len(df)
```

```python
df.head()
```

```python
df['petal length (cm)'].unique()
```

```python
def pred(x):
    if (x >= 1) & (x < 2.5) :
        return 0
    elif (x > 2.5) & (x < 4):
        return 1
    else:
        return 2
```

```python
df['pred_val'] = df['petal length (cm)'].apply(lambda x : pred(x))
```

```python
df[df.target != df.pred_val]
```

```python
total_error=df[df.target != df.pred_val].sample_weights.sum()# 0.26
```

```python
0.5* np.log((1-total_error)/total_error)
```

```python
alpha = 0.52
```

```python
df['new_weight'] = df.apply(lambda x : x.sample_weights+(np.exp(-alpha)) 
                            if (x.target == x.pred_val) else x.sample_weights+(np.exp(alpha)) , axis=1 )
```

```python
df[df.target == df.pred_val]
```

```python
df[df.target != df.pred_val]
```

```python
df.new_weight = df.new_weight/df.new_weight.sum()
```

# Build model using Adaboosting

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, RandomizedSearchCV , cross_val_score
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score , confusion_matrix , precision_score , recall_score
```

```python
data ,y = load_breast_cancer(return_X_y=True , as_frame=True)
```

```python
data
```

```python
y.values
```

```python
ada_clf = AdaBoostClassifier(n_estimators=200)
```

```python
param= {'n_estimators' : [50,100,150,200,300,400,500],
       'algorithm' :['SAMME', 'SAMME.R'] ,
       'learning_rate' : [1,2,5,10,50,100]}
```

```python
rnd_cv =RandomizedSearchCV(ada_clf,param_distributions= param , cv=10 , scoring='accuracy' ,
                           n_iter=100 , n_jobs=-1 ,random_state=12244)
```

```python
rnd_cv
```

```python
rnd_cv.fit(data,y)
```

```python
rnd_cv.best_estimator_
```

```python
rnd_cv.best_score_
```

```python
rnd_cv.best_params_
```

```python
cross_val_score(ada_clf,data,y,cv=10,scoring='accuracy')
```

```python
ada_clf.set_params(**rnd_cv.best_params_)
```

```python
ada_clf
```

```python
x_train,x_test,y_train,y_test = train_test_split(data,y,test_size=0.2,random_state=1223)
```

```python
ada_clf.fit(x_train,y_train)
```

```python
accuracy_score(y_test , ada_clf.predict(x_test))
```

```python
confusion_matrix(y_test , ada_clf.predict(x_test))
```

```python
precision_score(y_test , ada_clf.predict(x_test))
```

```python
recall_score(y_test , ada_clf.predict(x_test))
```
