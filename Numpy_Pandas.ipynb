```python
def decor(func):
    def inner(x):
        sum=func(x)##sum(lis)
        if sum%2==0:
            return 'Even'
        else:
            return 'Odd'
    return inner

def sum(x):
    s=0
    for i in x:
        s+=i
    return s
s1=decor(sum)
print(s1([2,3,4,56,21]))
```

```python
@decorate
def sum(x):
    s=0
    for i in x:
        s+=i
    return s
lis=[3,4,45,6]
sum(lis)

@decorate
def mul(x):
    m=1
    for i in x:
        m*=i
    return m
mul(lis)
```
