```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
cred_data = pd.read_csv(r'C:\Users\jiten\Downloads\crx.data', header=None , na_values='?')
```

```python
cred_data.info()
```

```python
cred_data.select_dtypes('O')
```

```python
cred_data.select_dtypes(['float64','int64'])
```

```python
cred_data = cred_data.iloc[:,[1,2,7,10,14]]
```

```python
cred_data.columns = ['x'+"_"+str(col) for col in cred_data.columns]
```

```python
cred_data.head()
```

```python
cred_data.x_1 = cred_data.x_1.astype('float64')
```

```python
cred_data.fillna(cred_data.x_1.mean(),axis=1,inplace=True)
```

```python
cred_data.isnull().sum()
```

```python
sc = StandardScaler()
cred_data = pd.DataFrame(sc.fit_transform(cred_data),columns=cred_data.columns)
```

```python
cred_data
```

```python
dbscan = DBSCAN(eps= 0.5 , min_samples= 3)
```

```python
label = dbscan.fit_predict(cred_data)
```

```python
cred_data.iloc[dbscan.core_sample_indices_,:]
```

```python
cred_data['label'] = label
```

```python
cred_data[cred_data.label == -1]
```
